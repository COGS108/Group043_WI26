## Background and Prior Work

Wine tasting has traditionally been an exclusive and highly specialized activity, requiring professional recognition that is done through formal training the completion of a sommelier exam. Evaluating wine quality involves numerous sensory factors, including texture, odor, flavor, color, and touch, all of which contribute to the overall perception of quality.<a name="cite_ref-1"></a><sup>1</sup> Traditionally, wine quality has been assessed subjectively through the human palette, where individual preferences and sensory perceptions dictate the quality. The human palette though, while fully capable of capturing the quality of wine, is a variable and in many ways subjective during the process. Therefore, objective methods have known to be used as well; for example, NIR spectroscopy(Near Infrared Spectroscopy) is used in evaluating certain compounds in flavor in the wine.<a name="cite_ref-1"></a><sup>2</sup>

This raises the question of whether we can understand more about how wine qualities are determined through studying its components. Specifically, how certain chemical components may influence the wine's taste or texture, and if whether these findings could help us predict a wine's quality given its statistics.

A similar study was conducted with two wine quality datasets, for red and white wine respectively. The study applied multiple regression to predict continuous quality scores, regression trees to capture nonlinear relationships, and classification models to group wines based on these properties. <a name="cite_ref-2"></a><sup>3</sup> They concluded that wine quality does not seem to have any major correlation with the chemical properties.

One prior study focused specifically on predicting red wine quality using physical and chemical properties like acidity, pH, alcohol content, etc. This work applied linear models and mulitnomial logistic regression to predict discrete quality ratings, demonstrating that objective wine characteristics can explain a decent portion of the variation and perceived quality in wine. Their results suggeste that seemingly simple statistical models can capture meaningful relationships between wine traits and quality.

Another study expanded on this idea except took a different approach in applying machine learning methods to identify which wine attributes and traits are most influential in determining overall quality. Rather than focusing on prediction alone, they emphasized feature importance, highlighting which chemical parameters contribute most to assessing quality. They found that certain variables like alcohol content and volatile acidity played a major role more times than not in predicting wine quality, which emphasized the idea that not all features contribute equally to the quality score.

Taking away from the studies, we aim to extend upon prior studies' results by investigating not only the features, but also whether certain combinations of chemical components may have a larger impact on the wine quality; specifically, if the  interactions within the features themselves could produce different results than just analyzing them one by one. While previous works determine that majority of the features do not contribute to the wine quality, by comparing models that study only individual predictors with models that have them interact, we hope to find whether accounting for these relationships improves predictive performance and would provide us a more comprehensive understanding of what are the contributing factors for wine quality.

## Hypothesis


We predict that wine quality is strongly determined by the combinations of the wine's properties rather than the properties on their own. As mentioned before, prior work has shown that the standalone properties such as alcohol content, acidity, etc. on their own tend to have strong influences with quality, while others do not. This suggests that wine quality can be deduced by a combination or balance between components rather than the significance of a feature on its own. We will examine this by exploring how combinations, such as sweetness/acidity, interact to influence the wine quality collectively. We believe that combinations such as sweetness/acidity in relation to other combinations such as sulfur/alcohol content or any other sulfur related combinations will be more predictive and decisive in terms of wine quality. The wine quality will be assessed with the dataset's already included quality score, with the traits of the wine (sugar, acidity), serving as independent variables as we explore whether modeling with interaction in mind among features will explain variation better than models which treat features separately.

* Dataset Name:  Wine Quality
* Link to the dataset: https://archive.ics.uci.edu/dataset/186/wine+quality
* Number of observations: 6497
* Number of variables: 12
* Input Variables (based on physicochemical tests):
  *   `fixed_acidity` (g(tartaric acid)/liter): Amount of non-volatile acids (e.g., tartaric acid, malic acit, etc.) in the wine. Contributes to the wine's overall acidity and taste.
  *   `volatile_acidity` (g(acetic acid)/liter): Amount of volatile acids, mainly acetic acid. High levels can cause an unpleasant vinegar-like taste.
  *   `citric_acid` (g/liter): Concentration of citric acid in the wine. It can add freshness and flavor, though it is usually present in small amounts.
  *   `residual_sugar`(g/liter): Amount of sugar remaining after fermentation stops. Higher values indicate sweeter wine.
  *   `chlorides` (g(sodium chloride)/liter): Amount of sodium chloride (salt) in the wine. High levels may negatively affect taste.
  *   `free_sulfur_dioxide` (mg/liter): Amount of unbound sulfur dioxide ($SO_2$) in the wine. It helps prevent microbial growth and oxidation.
  *   `total_sulfur_dioxide` (mg/liter): Total amount of sulfur dioxide ($SO_2$), including both free and bound. High levels can affect taste and aroma.
  *   `density` (g/cm³): Density of the wine. It is influenced by alcohol and sugar content.
  *   `pH`: Measure of acidity/alkalinity of the wine. Lower pH indicates higher acidity.
  *   `sulphates` (g(potassium sulphate)/liter): Amount of potassium sulphate. It contributes to microbial stability and may enhance flavor.
  *   `alcohol` (vol.%): Percentage of alcohol content in the wine. It strongly influences body and perception of quality.
* Output Variable (based on sensory data):
  *   `quality`: Sensory quality score, ranging from 0 (very bad) to 10 (excellent).

```python
################################### OUTLIERS ###################################

# define z-scores
numeric = df.select_dtypes(include="number").columns.tolist()
z_scores = pd.DataFrame(
    np.abs(stats.zscore(df[numeric])),
    columns=numeric,
    index=df.index,
)

# print % outliers in rows
for col in numeric:
    n = (z_scores[col] > 3).sum()
    print(f"{col:>20s}: {n:5d} outliers ({100 * n / len(df):1.1f}%)")

# flag outliers
df["outlier"] = (z_scores > 3).any(axis=1).map({True: "yes", False: "no"})
```

```python
df["outlier"]
```

**OUTLIERS:** We found the percentage of columns in each row of `df` that are more than 3 standard deviations from the mean, printed these values, and then flagged these columns with a new column `df["outlier"]` for potential investigation.

**CLEANLINESS:** As demonstrated in the writeup, all entries in this dataset have strict units and conform to those units. There is no missingness to address, and in the process of checking for outliers it is clear that no data is so wildly deviant from the mean so as to be the wrong datatype or unit, satisfying the definition of cleanliness.

#### Section 2  - Correlation structure

Next, we look at the distribution of quality scores across all features across the dataset to help us grasp a better understanding of the relationship between a wine's features and quality score. A heatmap helps identify potential associations between features, allowing us to quickly assess which variables may be more informative for predictive analysis while maintaining an overview of global data structure.

The features `sulphates` and `citric acid` follow `alcohol` in correlation strength. Although not as strong, these may also contribute to a higher wine quality.  Other features show weaker correlations (closer to 0), indicating they have smaller linear associations with the target variable. On the other hand, `Density` and `volatile acidity` are the features with the strongest negative correlation to `quality`, suggesting that higher acidity levels and higher density are associated with lower quality wines. #### Section 3 of EDA  - Feature Distribution Analysis?

To further take a look at the relationship between alcohol and quality, we use a histogram to observe the distribution of alcohol across quality scores. To improve visualization clarity, wine quality scores were grouped into three categories: low(0-4), medium(5-6), and high quality(7-9). 

```python
def quality_group(q):
    if q <= 4:
        return "Low"
    elif q <= 6:
        return "Medium"
    else:
        return "High"

df["quality_group"] = df["quality"].apply(quality_group)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="alcohol",
    hue="quality_group",
    bins=30,
    kde=True
)

plt.xlabel("Alcohol (% vol)")
plt.ylabel("Count")
plt.title("Alcohol Distribution Across Quality Groups")
plt.show()
```

The histogram suggests a slight shift toward higher alcohol content for higher quality wines. Although there is substantial overlap among quality groups, there appears to be some separation in the distribution, supporting the observation that alcohol has a moderate positive correlation with wine quality.

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="density",
    hue="quality_group",
    bins=30,
    kde=True
)

plt.xlabel("Density (g/cm³)")
plt.ylabel("Count")
plt.title("Density Distribution Across Quality Groups")
plt.show()
```

The histogram shows complete overlap among all quality groups for density, indicating that density alone does not effectively distinguish wine quality levels. This observation is consistent with the weak correlation between density and quality identified.

This boxplot shows the distribution of alcohol content across wine quality ratings. Each box represents the interquartile range (IQR), with the median displayed as a line inside the box and whiskers indicating the spread of the data.

```python
sns.boxplot(data=df, x='quality', y='alcohol')
plt.title("Alcohol vs Wine Quality")
plt.show()
```

The plot shows an **upward trend**, where higher quality wines tend to have a higher alcohol content. The median alcohol level, as denoted by the horizontal line, increases from lower quality wines to higher quality wines. This suggests that alcohol content is a good/strong predictor of wine quality for this dataset.

There appears to be a very slight upward trend, where the median sulphate level rises, particularly between quality levels 5 and 7. Also notable are the several outliers with higher sulphate values that are present, especially in mid-quality wines, showing a high variability in sulphate content for this group. Through exploratory data analysis and visualization, we examined how several chemical attributes such as alcohol content, density, and volatile aciditiy relate to the quality scores given to the wines. One observation is that most wines are concentrated around quality scores of 5 and 6, with very few distributing among very low or very high scores. As the dataset is skewed towards more average scores, it may be difficult for us to distinguish features that separate high-quality wine from low-quality wine.

Across various visualizations and analysis, we find that alcohol content consistently displayed the strongest relationship with wine quality. Although there was substantial overlap between quality groups, the overall upward trend suggests that alcohol content is one of the more informative predictors of wine quality in this dataset.

In contrast, volatile acidity showed a negative relationship with quality.
